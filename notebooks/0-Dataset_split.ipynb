```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv('..\\data\\raw\\credit_risk_dataset.csv')
```

```python
# 1. Separa a variável alvo (y) das variáveis preditoras (X)
X = df.drop(columns=['loan_status'])
y = df['loan_status']

# Divide o dataset original em 70% para treino e 30% para o "resto"
X_train, X_resto, y_train, y_resto = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=y
)

# Pega nos 30% do "resto" e divide exatamente ao meio (50% para cada lado)
# Isto vai gerar 15% para validação e 15% para teste em relação ao dataset original
X_val, X_test, y_val, y_test = train_test_split(
    X_resto, y_resto, test_size=0.50, random_state=42, stratify=y_resto
)
```

```python
dftreino = pd.concat([X_train, y_train], axis=1)
dftreino.to_csv('..\\data\\raw\\df_train.csv', index=False)
```

```python
df_teste = pd.concat([X_test, y_test], axis=1)
df_teste.to_csv('..\\data\\raw\\df_test.csv', index=False)
```

```python
df_validacao = pd.concat([X_val, y_val], axis=1)
df_validacao.to_csv('..\\data\\raw\\df_validation.csv', index=False)
```
